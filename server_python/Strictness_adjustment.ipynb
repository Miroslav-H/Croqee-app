```python
img1 = np.array(Image.open(dir_path + '/test_images/geometrical5.png'))
img2 = np.array(Image.open(dir_path + '/test_images/geometrical1.png'))
```

```python
def get_strictness_parameter(img, max_dist=50):
    s = 2 * max_dist + 1
    f = cv.getStructuringElement(cv.MORPH_ELLIPSE,(s, s))
    binary_img = (img > 100).astype(np.uint8)
    dst = cv.distanceTransform(binary_img.astype(np.uint8), cv.DIST_L2, 3)
    covering = cv.dilate(1-binary_img, f)
    return np.sum(dst * covering) / np.sum(covering)
    
```

```python
s = 2 * 50 + 1
f = cv.getStructuringElement(cv.MORPH_ELLIPSE,(s, s))
binary_img = (img2 > 100).astype(np.uint8)
dst = cv.distanceTransform(binary_img.astype(np.uint8), cv.DIST_L2, 3)
covering = cv.dilate(1-binary_img, f)

plt.figure(figsize=(16,10))
plt.imshow(dst)
plt.figure(figsize=(16,10))
plt.imshow(covering)
```
